```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import scipy
```

```python
def beta_dist(alpha=2.0, beta=2.0):
    mean = alpha/(alpha+beta)
    median = (alpha-(1/3))/(alpha+beta-(2/3))
    mode = (alpha-1)/(alpha+beta-2)

    x = np.linspace(0,1,100)
    pdf = scipy.stats.beta.pdf(x, alpha, beta)

    plt.figure(figsize=(10,5))
    plt.plot(x, pdf, linewidth=3, label='Beta PDF',
            color='navy')
    plt.fill_between(x, pdf, alpha=0.25,color='lightsteelblue')

    plt.axvline(mean, linestyle='--', linewidth=2,label=f"Mean = {mean:.3f}", 
                color='royalblue')
    plt.axvline(median, linestyle='-.', linewidth=2,label=f"Median = {median:.3f}", 
                color='cornflowerblue')
    plt.axvline(mode, linestyle=':', linewidth=3,label=f"Mode = {mode:.3f}",
               color='lightsteelblue')

    plt.title("Measures of Centering",fontsize=20)

    plt.xlabel("Input Value")
    plt.ylabel(f"Beta Distribution (α={alpha:.1f}, β={beta:.1f})")
    plt.xlim(0,1)
    plt.ylim(0,4)
    plt.grid(alpha=0.3)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()
    
interact(beta_dist,
        alpha=FloatSlider(value=2,
                          min=2,
                          max=8,
                          step=0.5,
                          description='α'
                        ),
        beta=FloatSlider(value=2,
                          min=2,
                          max=8,
                          step=0.5,
                          description='β'
                        ))
```

```python
def normal_dist(sigma=1):
    mean = 0
    x = np.linspace(mean - 4*sigma, mean + 4*sigma, 500)
    y = scipy.stats.norm.pdf(x, mean, sigma)
    
    variance = sigma**2
    std = sigma
    mad = sigma*np.sqrt(2)*scipy.special.erfinv(.5)

    plt.figure(figsize=(10,5))
    plt.plot(x, y, label="Theoretical PDF",
            color = 'purple')
    plt.axvline(mean, linewidth=2,label=f"Mean = {mean:.2f}", color='purple')
    
    plt.vlines([mean+variance,mean-variance], 0, 2,
                 linestyle="-.",linewidth=2,
               label=f"Variance = {variance:.2f}",
                    color='m')
    plt.vlines([mean+sigma, mean-sigma], 0, 2,
                 linestyle="--",linewidth=3,
               label=f"Stand. Dev. = {std:.2f}",
                    color='violet')
    plt.vlines([mean+mad, mean-mad], 0, 2,
                 linestyle=":",linewidth=3,
               label=f"Mean Abs. Dev. = {mad:.2f}",
                    color='plum')
    plt.fill_between(x[(mean+variance>x)& (x>mean-variance)],
                     y[(mean+variance>x)& (x>mean-variance)],
                     alpha=0.25, 
                    color ='m')
    plt.fill_between(x[(mean+sigma>x)& (x>mean-sigma)],
                     y[(mean+sigma>x)& (x>mean-sigma)],
                     alpha=0.25, 
                    color ='violet')
    plt.fill_between(x[(mean+mad>x)& (x>mean-mad)],
                     y[(mean+mad>x)& (x>mean-mad)],
                     alpha=0.25, 
                    color ='plum')
    plt.fill_between(x, y, alpha=0.25,
                    color='thistle')

    plt.title("Measures of Variation",fontsize=20)
    plt.xlabel("Input Value")
    plt.ylabel(f"Normal Distribution (μ={mean:.1f}, σ={sigma:.1f})")
    plt.xlim(-1,1)
    plt.ylim(0,2)
    plt.grid(alpha=0.3)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

interact(normal_dist, sigma=FloatSlider(value=.2,min=0.2,max=1,step=0.1,description='σ'))
```
